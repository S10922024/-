```
stop_words <- c("在", "之", "更", "個", "未", "而", "也", "仍","上","下","的","與","前","中","是","只","若","和","從","於","會","後","台","將","對","該")
writeLines(stop_words, "stop_words.txt")
```

```
cutter <- worker(user = "new_words.txt", stop_word = "stop_words.txt", bylines = FALSE)
seg_words <- cutter[content]
seg_words
```
