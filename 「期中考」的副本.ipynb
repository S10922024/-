```
install.packages("SportsAnalytics")#安裝套件
```

```
library(SportsAnalytics)#載入套件

NBA1516<-fetch_NBAPlayerStatistics("15-16")#讀入資料
```

```
str(NBA1516)
```

```
NBA15160rder<-NBA1516[order(NBA1516$Blocks,decreasing = T),]#蓋火鍋最多的前20名選手
NBA15160rder[1:20,]
```

```
subset(NBA1516,Blocks  =="25")#取蓋火鍋剛好等於25的選手
```

```
NBA1516[grepl("Anthony",NBA1516$Name),]#找姓名中有Anthony的人
```

```
install.packages("ggplot2")#安裝ggplot2
```

```
library(ggplot2)#載入套件
```

```
qplot( ThreesMade , ThreesAttempted, data = NBA1516, color = Team)#每隊3分的得分散佈圖(x=三分球進球分數,y=三分球出手次數)
```

```
qplot( ThreesMade, data = NBA1516, facets = Position ~ ., binwidth = 18)#每個位置得三分的分數
```

```
ggplot() + geom_bar(data = NBA1516,aes(x=Team, y=ThreesMade), stat = "identity")#每隊得到三分球的分數(x=隊伍名稱, y=三分球得分數)
```

```
ggplot( data = NBA1516, aes(x = TotalRebounds , y = OffensiveRebounds , color = Position )) + geom_line() #每個位置對總籃板和進攻籃板的關係(x=總籃板, y=進攻籃板)
```

```
ggplot( data = NBA1516, aes(x= Position, y = Blocks )) + geom_boxplot()#每個位置蓋火鍋的盒狀圖(x=位置, y=蓋火鍋數)
```
